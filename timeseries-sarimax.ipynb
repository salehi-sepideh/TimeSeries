```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/kaggle/input/timeseriessarimax/jshare.csv')
data.head()
```

```python
plt.figure(figsize=(20,6))
plt.plot(data['date'] , data['data'])
plt.show()
```

```python
#find best p,q,ps,qs
from itertools import product
p=range(0,4,1)
d=1
q=range(0,4,1)
ps=range(0,4,1)
ds=1
qs=range(0,4,1)
s=4
params=list(product(p,q,ps,qs))
params
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
results=[]
for order in params:
    try:
        print(order)
        model=SARIMAX(data['data'] , order=(order[0] , d , order[1]),
                      seasonal_order=(order[2] , ds , order[3],s))
        res=model.fit()
        results.append([order,res.aic])
    except:
        continue
```

```python
#best AIc
r=pd.DataFrame(results)
r.columns=('order', 'AIC')
r.sort_values(by='AIC' , ascending=True)
```

```python
model = SARIMAX(data['data'], order=(2, 1, 3), seasonal_order=(0, 1, 0, 4))
res = model.fit()
res.summary()
```

```python
data['sarimax']=res.fittedvalues
data['sarimax'][:5]= np.NaN
forecast=res.predict(start=len(data) , end=len(data)+8)
forecast=pd.concat([data['sarimax'],forecast])
```

```python
plt.figure(figsize=(20,6))
plt.plot(data['data'])
plt.plot(forecast)
plt.show()
```
